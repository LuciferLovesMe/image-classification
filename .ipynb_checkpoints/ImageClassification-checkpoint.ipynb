### Data Loading

#### Menampilkan resolusi gambar pada dataset

```python
def checkResolution (path) :
    resolutions = []

    for classes_name in os.listdir(path):
        class_path = os.path.join(path, classes_name)
        if os.path.isdir(class_path):
            for imgs_name in os.listdir(class_path):
                img_path = os.path.join(class_path, imgs_name)
                try:
                    with Image.open(img_path) as img:
                        resolutions.append(img.size)
                except:
                    print("Gambar tidak valid.")
                    continue

    count_resolutions = Counter(resolutions)

    for resolution, counter in count_resolutions.items():
        print(f'\nResolusi {resolution}: {counter} Gambar')
        
    resolutions_width = [resolution[0] for resolution, counter in count_resolutions.items()]
    resolutions_height = [resolution[1] for resolution, counter in count_resolutions.items()]

    plt.figure(figsize=(12, 8))
    plt.scatter(resolutions_width, resolutions_height, alpha=0.5)
    plt.title("Scatter Distribusi resolusi gambar")
    plt.xlabel("Lebar (width)")
    plt.ylabel("Tinggi (height)")
    plt.show()
```

```python
checkResolution(path)
```

```python
# Resize resolusi menjadi seragam
new_path = 'resized_dataset'
target_size = (136, 102)

for classes_name in os.listdir(path):
    class_path = os.path.join(path, classes_name)

    new_classes_path = os.path.join(new_path, classes_name)
    os.makedirs(new_classes_path, exist_ok=True)
    
    if os.path.isdir(class_path):
        for imgs_name in os.listdir(class_path):
            img_path = os.path.join(class_path, imgs_name)
            try:
                with Image.open(img_path) as img:
                    img_resized = img.resize(target_size)
                    img_resized.save(os.path.join(new_classes_path, imgs_name))
            except Exception as e:
                print(f"Gambar tidak valid. {imgs_name}")
                continue
```

### Data Preprocessing

#### Split Dataset

```python
dataset_path = 'resized_dataset'
base_splitted_path = 'splitted_data'
train_path = os.path.join(base_splitted_path, 'train_set')
test_path = os.path.join(base_splitted_path, 'test_set')
val_path = os.path.join(base_splitted_path,'val_set')

for directory in [train_path, test_path, val_path]:
    os.makedirs(directory, exist_ok=True)

class_list = os.listdir(dataset_path)

for class_name in class_list:
    class_path = os.path.join(dataset_path, class_name)
    
    if not os.path.isdir(class_path):
        continue

    class_train_path = os.path.join(train_path, class_name)
    class_test_path = os.path.join(test_path, class_name)
    class_val_path = os.path.join(val_path, class_name)
    for class_splitted_path in [class_train_path, class_test_path, class_val_path]:
        os.makedirs(class_splitted_path, exist_ok=True)
        
    imgs_name = os.listdir(class_path)

    train_files, testing_val_files = train_test_split(imgs_name, test_size=0.2, random_state=42)
    val_files, test_files = train_test_split(testing_val_files, test_size=0.5, random_state=42)

    def putSplittedFiles (files, source_path, target_path):
        for file in files:
            shutil.copy(os.path.join(source_path, file), os.path.join(target_path, file))

    putSplittedFiles(train_files, class_path, os.path.join(train_path, class_name))
    putSplittedFiles(test_files, class_path, os.path.join(test_path, class_name))
    putSplittedFiles(val_files, class_path, os.path.join(val_path, class_name))
```

```python
checkResolution(base_splitted_path)
```

#### Data image loading

```python
train_datagenerator = ImageDataGenerator(rescale=1./255)
testval_datagenerator = ImageDataGenerator(rescale=1./255)
```

```python
# Load data training
train_image_generator = train_datagenerator.flow_from_directory(
    train_path,
    target_size=(136, 102),
    batch_size=32,
    class_mode='categorical'
)

# Load data testing
test_image_generator = testval_datagenerator.flow_from_directory(
    test_path,
    target_size=(136, 102),
    batch_size=32,
    class_mode='categorical'
)

# Load data validation
val_image_generator = testval_datagenerator.flow_from_directory(
    val_path,
    target_size=(136, 102),
    batch_size=32,
    class_mode='categorical'
)
```

## Modelling

### Modeling Menggunakan CNN

```python
cnn = models.Sequential()

cnn.add(layers.Conv2D(64, (3, 3), activation='relu', kernel_initializer=he_normal(), input_shape=(136, 102, 3))),
cnn.add(layers.BatchNormalization())
cnn.add(layers.MaxPooling2D(pool_size=(2, 2)))

cnn.add(layers.Conv2D(32, (3, 3), activation='relu', kernel_initializer=he_normal(), input_shape=(136, 102, 3))),
cnn.add(layers.BatchNormalization())
cnn.add(layers.MaxPooling2D(pool_size=(2, 2)))
cnn.add(layers.Dropout(0.2)) 

cnn.add(layers.Conv2D(128, (3, 3), activation='relu', kernel_initializer=he_normal(), input_shape=(136, 102, 3))),
cnn.add(layers.BatchNormalization())
cnn.add(layers.MaxPooling2D(pool_size=(2, 2)))
cnn.add(layers.Dropout(0.3)) 

cnn.add(layers.Flatten())
cnn.add(layers.Dense(128, activation='relu', kernel_initializer=he_normal()))
cnn.add(layers.Dropout(0.4))
cnn.add(layers.Dense(3, activation='softmax', kernel_initializer=he_normal()))

adam_optimizer = Adam(learning_rate=1e-4)
cnn.compile(optimizer=adam_optimizer, loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
cnn.summary()
```

```python
early_stopping = EarlyStopping(
    min_delta=0.001,
    patience=10,
    restore_best_weights=True
)
```

```python
model_checkpoint = ModelCheckpoint(
    'best_model.h5',
    save_best_only=True,
    save_weights_only=False,
    verbose=1
)
```

```python
log_path = os.path.join("logs", "fit")
os.makedirs(log_path, exist_ok=True)

tensorboard = TensorBoard(
    log_dir=log_path,
    histogram_freq=1
)
```

```python
reduce_lr = ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.1,
    patience=3,
)
```

```python
callbacks = [early_stopping, model_checkpoint, reduce_lr, tensorboard]
```

```python
history_cnn = cnn.fit(
    train_image_generator,
    steps_per_epoch=train_image_generator.samples // train_image_generator.batch_size,
    epochs=30,
    validation_data=val_image_generator,
    validation_steps=val_image_generator.samples // val_image_generator.batch_size,
    callbacks=callbacks
)
```

## Evaluasi dan Visualisasi

```python
test_loss, test_acc = cnn.evaluate(test_image_generator, steps=test_image_generator.samples // test_image_generator.batch_size)
print(f'Test Loss: {test_loss}')
print(f'Test Accuracy: {test_acc}')
```

```python
val_loss, val_acc = cnn.evaluate(val_image_generator, steps=val_image_generator.samples // val_image_generator.batch_size)
print(f'Validation Loss: {val_loss}')
print(f'Validation Accuracy: {val_acc}')
```

## Inference (Optional)

```python
tflite_model_path = 'tflite_model/model.tflite'
interpreter_model = tf.lite.Interpreter(model_path=tflite_model_path)
interpreter_model.allocate_tensors()
```

```python
input_new_image = interpreter_model.get_input_details()
output_new_image = interpreter_model.get_output_details()
_, height, width, channels = input_new_image[0]['shape']

class_labels = ["Boot", "Shoe", "Sandal"]
```

```python
def preprocessing_new_image (image_path):
    new_img = Image.open(image_path).convert("RGB").resize((width, height))
    img_array = np.array(new_img)
    img_array = img_array / 255.0
    img_array = np.expand_dims(img_array, axis=0).astype(np.float32)
    return img_array, new_img
```

```python
def predict_new_image (interpreter, image_path):
    new_image, display_image = preprocessing_new_image(image_path)
    interpreter.set_tensor(input_new_image[0]['index'], new_image)

    interpreter.invoke()
    output_data = interpreter.get_tensor(output_new_image[0]['index'])
    
    predicted = np.argmax(output_data)
    predicted_class = class_labels[predicted]
    predicted_accuracy = output_data[0][predicted] * 100

    plt.imshow(display_image)
    plt.title(f'Prediksi Kelas {predicted_class}, Akurasi: {predicted_accuracy}%')
    plt.axis("off")
    plt.show()
    
    return predicted_class, predicted_accuracy
```

```python
predict_new_image(interpreter_model, 'sample-sandal.jpg')
```

```python
predict_new_image(interpreter_model, 'sample-boot.jpg')
```
